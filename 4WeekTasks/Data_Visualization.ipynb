Promt:beri kode untuk menampilkan bar chart dari fitur categorical

```python
import matplotlib.pyplot as plt

# Menghitung jumlah masing-masing kategori dalam fitur 'cut'
cut_counts = df['cut'].value_counts()

# Membuat bar chart
plt.figure(figsize=(8, 6))
cut_counts.plot(kind='bar', color='skyblue')
plt.title('Jumlah Kategori pada Fitur "cut"')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)  # Untuk mengatur label sumbu x tidak miring
plt.show()
```

Promt:tampilkan distribusi numerical features dengan 2 baris 3 kolom

```python
import matplotlib.pyplot as plt
import math

# Daftar fitur numerik yang ingin Anda tampilkan
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Menghitung jumlah baris dan kolom subplot
num_rows = 2
num_cols = 3

# Menghitung jumlah total subplot
num_subplots = num_rows * num_cols

# Membuat subplot
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Loop melalui setiap fitur numerik dan membuat histogram
for i, feature in enumerate(numerical_features):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]

    ax.hist(df[feature], bins=30, color='skyblue')
    ax.set_title(f'Distribusi dari Fitur "{feature}"')
    ax.set_xlabel(feature)
    ax.set_ylabel('Jumlah')

# Menghilangkan subplot yang tidak terpakai
for i in range(len(numerical_features), num_subplots):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row, col])

# Memastikan layout subplot rapi
plt.tight_layout()

# Menampilkan subplot
plt.show()
```

Promt:beri kode untuk menampilkan boxplot masing-masing features dari numerical features dengan 2 baris 3 kolom, dan boxplot menghadap horizontal

```python
import matplotlib.pyplot as plt
import math

# Daftar fitur numerik yang ingin Anda tampilkan
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Menghitung jumlah baris dan kolom subplot
num_rows = 2
num_cols = 3

# Membuat subplot
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Loop melalui setiap fitur numerik dan membuat boxplot horizontal
for i, feature in enumerate(numerical_features):
    row = i // num_cols
    col = i % num_cols

    # Membuat boxplot horizontal
    axes[row, col].boxplot(df[feature], vert=False)
    axes[row, col].set_title(f'Boxplot {feature}')
    axes[row, col].set_xlabel(feature)

# Menghilangkan subplot yang tidak terpakai
for i in range(len(numerical_features), num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    fig.delaxes(axes[row, col])

# Memastikan layout subplot rapi
plt.tight_layout()

# Menampilkan subplot
plt.show()
```
